```python
df_limpio["MaritalStatus"].value_counts()
```

```python
df_limpio["MaritalStatus"] = df_limpio["MaritalStatus"].str.title()

df_limpio["MaritalStatus"] = df_limpio["MaritalStatus"].str.replace("Marreid", "Married")
```

```python
df_limpio["DistanceFromHome"] = df_limpio["DistanceFromHome"].abs()
df_limpio
```

```python

df_limpio['EnvironmentSatisfaction'].str.split(expand=True).get([0])
```

```python
df_limpio["RemoteWork"] = df_limpio["RemoteWork"].str.replace("1", "True").replace("0", "False").replace("Yes", "True")
df_limpio
```
